# Question 3 soloution

## 1. Probability Space

A suitable probability space is
$$
\Omega=\{0,1,2,3\}, \qquad \mathfrak F=\mathfrak P(\Omega),
$$
where $\mathfrak P(\Omega)$ denotes the power set of $\Omega$.

Define the probability measure on singletons as
$$
P(0)=0.50,\quad P(1)=0.30,\quad P(2)=0.15,\quad P(3)=0.05.
$$

For any event $A \subseteq \Omega$,
$$
P(A)=\sum_{\omega \in A} P(\omega).
$$

---

## 2. Definition of the Random Variable

Define $X:\Omega \to \mathbb{R}$ by
$$
X(\omega)=\omega, \quad \omega \in \Omega.
$$

Thus, $X$ directly represents the number of failures.

---

## 3. Probability Mass Function

The probability mass function of $X$ is
$$
p_X(x)=P(X=x)=
\begin{cases}
0.50, & x=0,\\
0.30, & x=1,\\
0.15, & x=2,\\
0.05, & x=3,\\
0, & \text{otherwise}.
\end{cases}
$$

---

## 4. Expectation

The expected value is
$$
E[X]=\sum x\,p_X(x)
=0(0.50)+1(0.30)+2(0.15)+3(0.05)=0.75.
$$

---

## 5. Variance

First compute
$$
E[X^2]=0^2(0.50)+1^2(0.30)+2^2(0.15)+3^2(0.05)=1.35.
$$

Then
$$
\mathrm{Var}(X)=E[X^2]-(E[X])^2
=1.35-(0.75)^2
=0.7875.
$$

---

## 6. Required Probabilities

$$
P(X \ge 1)=0.30+0.15+0.05=0.50,
$$

$$
P(X \ge 2)=0.15+0.05=0.20.
$$

---

## 7. Interpretation

The expectation $E[X]=0.75$ represents the **average number of transformer failures per day**.

The variance $\mathrm{Var}(X)=0.7875$ measures the **variability in daily failures**, indicating how much the number of failures fluctuates around the average.

A higher variance would imply more unpredictable daily behavior, while a lower variance would indicate more consistent operation.

# Question 4 soloution

## 1. Probability Space

A suitable probability space is
$$
\Omega=[0,1], \qquad \mathfrak F=\mathfrak B([0,1]),
$$
where $\mathfrak B([0,1])$ is the Borel $\sigma$-algebra.

Define the probability measure by
$$
P(A)=\int_A f(x)\,dx, \quad A\in \mathfrak F,
$$
where
$$
f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x(1-x)^{17}, & 0<x<1,\\
0, & \text{otherwise}.
\end{cases}
$$

---

## 2. Definition of the Random Variable

Define $X:\Omega \to \mathbb{R}$ by
$$
X(\omega)=\omega, \quad \omega\in[0,1].
$$

Thus, $X$ represents the defective proportion in a batch.

---

## 3. Probability Density Function

The density of $X$ is
$$
f_X(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x(1-x)^{17}, & 0<x<1,\\
0, & \text{otherwise}.
\end{cases}
$$

---

## 4. Verification of Valid Density

- Clearly, $f(x)\ge 0$ for all $x$  
- Also,
$$
\int_{-\infty}^{\infty} f(x)\,dx
=\int_0^1 \frac{1}{B(2,18)}x(1-x)^{17}\,dx
=\frac{B(2,18)}{B(2,18)}=1
$$

Hence, $f$ is a valid probability density function.

---

## 5. Expectation

For a Beta$(\alpha,\beta)$ distribution,
$$
E[X]=\frac{\alpha}{\alpha+\beta}.
$$

Thus,
$$
E[X]=\frac{2}{2+18}=0.10.
$$

---

## 6. Variance

The variance is given by
$$
\mathrm{Var}(X)=\frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)}.
$$

Hence,
$$
\mathrm{Var}(X)=\frac{2\cdot 18}{20^2 \cdot 21}
=\frac{36}{8400}
=\frac{3}{700}
\approx 0.00429.
$$

---

## 7. Probability $P(X \le 0.10)$

$$
P(X\le 0.10)=\int_0^{0.10} \frac{1}{B(2,18)}x(1-x)^{17}\,dx.
$$

Equivalently,
$$
P(X\le 0.10)=I_{0.10}(2,18),
$$
where $I_x(\alpha,\beta)$ is the regularized incomplete Beta function.

---

## 8. Probability $P(0.05 \le X \le 0.15)$

$$
P(0.05\le X\le 0.15)=\int_{0.05}^{0.15} \frac{1}{B(2,18)}x(1-x)^{17}\,dx.
$$

Equivalently,
$$
P(0.05\le X\le 0.15)=I_{0.15}(2,18)-I_{0.05}(2,18).
$$

---

## 9. Why Beta Distribution is Appropriate

The Beta distribution is well-suited here because:
- The variable represents a **proportion**, bounded between 0 and 1  
- The Beta family is flexible and can model different shapes of distributions over this interval  
- It is commonly used for modeling probabilities and rates  

---

## 10. Interpretation

The expectation
$$
E[X]=0.10
$$
indicates that, on average, about **10% of items are defective**.

The variance
$$
\mathrm{Var}(X)\approx 0.00429
$$
measures how much the defective proportion varies between batches.

A small variance suggests that the process is relatively **stable**, while a larger variance would indicate greater inconsistency in product quality.

# Question 5 soloution

## 1. Support of $X$

Since the installation time ranges between the minimum and maximum estimates,
$$
4 \leq X \leq 10.
$$

---

## 2. Probability Density Function

For $X \sim \mathrm{Triangular}(a=4, b=10, c=6)$, the density is

$$
f_X(x)=
\begin{cases}
\dfrac{2(x-4)}{(10-4)(6-4)}, & 4 \leq x < 6,\\
\dfrac{2(10-x)}{(10-4)(10-6)}, & 6 \leq x \leq 10,\\
0, & \text{otherwise}.
\end{cases}
$$

Simplifying,
$$
f_X(x)=
\begin{cases}
\dfrac{x-4}{6}, & 4 \leq x < 6,\\
\dfrac{10-x}{12}, & 6 \leq x \leq 10,\\
0, & \text{otherwise}.
\end{cases}
$$

---

## 3. Verification of Density

We check that
$$
\int_{4}^{10} f_X(x)\,dx = 1.
$$

Compute:
$$
\int_{4}^{6} \frac{x-4}{6}\,dx = \frac{1}{6}\int_{4}^{6}(x-4)\,dx = \frac{1}{3},
$$

$$
\int_{6}^{10} \frac{10-x}{12}\,dx = \frac{1}{12}\int_{6}^{10}(10-x)\,dx = \frac{2}{3}.
$$

Thus,
$$
\int_4^{10} f_X(x)\,dx = \frac{1}{3} + \frac{2}{3} = 1.
$$

So $f_X$ is a valid probability density function.

---

## 4. Expectation

For a triangular distribution,
$$
E[X]=\frac{a+b+c}{3}.
$$

Thus,
$$
E[X]=\frac{4+10+6}{3}=\frac{20}{3}\approx 6.67.
$$

---

## 5. Variance

$$
\mathrm{Var}(X)=\frac{a^2+b^2+c^2-ab-ac-bc}{18}.
$$

Substituting values:
$$
\mathrm{Var}(X)=\frac{16+100+36-40-24-60}{18}
=\frac{28}{18}
=\frac{14}{9}\approx 1.56.
$$

---

## 6. Interpretation of Parameters

- $a=4$: minimum possible installation time  
- $b=10$: maximum possible installation time  
- $c=6$: most likely (mode) installation time  

---

## 7. Why Triangular Distribution?

The triangular distribution is appropriate when:
- limited historical data is available  
- experts can estimate minimum, maximum, and most likely values  
- the variable is bounded  

Thus, it provides a simple yet practical model for uncertainty.

---

## 8. Probability $P(X \leq 5)$

$$
P(X \leq 5)=\int_{4}^{5} \frac{x-4}{6}\,dx
$$

$$
= \frac{1}{6}\int_{4}^{5}(x-4)\,dx
= \frac{1}{6}\cdot \frac{1}{2}
= \frac{1}{12}.
$$

So,
$$
P(X \leq 5)=\frac{1}{12}\approx 0.0833.
$$

---

## 9. Probability $P(5 \leq X \leq 8)$

Split at $x=6$:
$$
P(5 \leq X \leq 8)=\int_{5}^{6} \frac{x-4}{6}\,dx + \int_{6}^{8} \frac{10-x}{12}\,dx.
$$

First part:
$$
\frac{1}{6}\int_{5}^{6}(x-4)\,dx = \frac{1}{4}.
$$

Second part:
$$
\frac{1}{12}\int_{6}^{8}(10-x)\,dx = \frac{1}{2}.
$$

Thus,
$$
P(5 \leq X \leq 8)=\frac{1}{4}+\frac{1}{2}=\frac{3}{4}.
$$

---

## 10. Practical Use

This model helps in:

- Project planning: estimating the likelihood of completing installation within a given time  
- Risk analysis: understanding uncertainty in completion time  
- Sensitivity analysis: evaluating how delays affect schedules, staffing, and contingency plans  

It provides a simple way to incorporate expert judgment into decision-making.

# Question 6 soloution

## 1. Shape Parameters

Given
$$
a=8, \quad b=20, \quad c=11,
$$

the PERT parameters are
$$
\alpha = 1 + 4\left(\frac{c-a}{b-a}\right)
= 1 + 4\left(\frac{3}{12}\right)
= 2,
$$

$$
\beta = 1 + 4\left(\frac{b-c}{b-a}\right)
= 1 + 4\left(\frac{9}{12}\right)
= 4.
$$

---

## 2. Support

Since $X$ represents completion time,
$$
8 \leq X \leq 20.
$$

---

## 3. Probability Density Function

$$
f_X(x)=\frac{(x-8)^{1}(20-x)^{3}}{B(2,4)\,12^5}, \quad 8 \leq x \leq 20,
$$

and $f_X(x)=0$ otherwise.

Using
$$
B(2,4)=\frac{1}{20},
$$

we obtain
$$
f_X(x)=\frac{20(x-8)(20-x)^3}{12^5}, \quad 8 \leq x \leq 20.
$$

---

## 4. Probability Space

$$
\Omega=[8,20], \quad \mathfrak F=\mathfrak B([8,20])
$$

$$
P(A)=\int_A f_X(x)\,dx
$$

---

## 5. Definition of $X$

$$
X(\omega)=\omega, \quad \omega \in [8,20].
$$

Thus, $X$ represents the task duration.

---

## 6. Expectation

For the PERT distribution,
$$
E[X]=\frac{a+4c+b}{6}.
$$

Substituting,
$$
E[X]=\frac{8+4(11)+20}{6}=\frac{72}{6}=12.
$$

---

## 7. Variance (Approximation)

$$
\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7}
$$

$$
=\frac{(12-8)(20-12)}{7}
=\frac{32}{7}\approx 4.57.
$$

---

## 8. Standard Deviation

$$
\sigma=\sqrt{\frac{32}{7}}\approx 2.14.
$$

Rule-of-thumb check:
$$
\sigma \approx \frac{b-a}{6}=\frac{12}{6}=2,
$$
which is close.

---

## 9. Comparison with Triangular Model

| Quantity           | Triangular | PERT |
|------------------|-----------:|-----:|
| Mean             | 13         | 12   |
| Variance         | 8.17       | 4.57 |
| Std. deviation   | 2.86       | 2.14 |
| Shape            | Linear     | Smooth |

**Key insight:**  
The PERT model gives:
- a mean closer to the most likely value  
- lower variance (less extreme uncertainty)  
- smoother, more realistic distribution  

---

## 10. Interpretation

- $a=8$: minimum (optimistic time)  
- $b=20$: maximum (pessimistic time)  
- $c=11$: most likely time  
- $E[X]=12$: expected completion time  
- $\sigma \approx 2.14$: variability in duration

The PERT approach is particularly useful when decisions rely on expert estimates rather than large datasets.

# Question 7 soloution

## 1. Determining $b$

The PERT mean is given by
$$
E[X]=\frac{a+4c+b}{6}.
$$

We are told that the expected value equals the most likely value, so
$$
E[X]=c.
$$

Substituting,
$$
\frac{a+4c+b}{6}=c.
$$

Multiplying by $6$,
$$
a+4c+b=6c.
$$

Rearranging,
$$
a+b=2c.
$$

Hence,
$$
b=2c-a.
$$

Substituting $a=5$ and $c=11$,
$$
b=2(11)-5=22-5=17.
$$

---

## 2. Effect on the Variance

Using the approximation
$$
\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7},
$$

and since $E[X]=c$,
$$
\mathrm{Var}(X)=\frac{(c-a)(b-c)}{7}.
$$

Now substitute $b=2c-a$:
$$
b-c=(2c-a)-c=c-a.
$$

Thus,
$$
\mathrm{Var}(X)=\frac{(c-a)(c-a)}{7}
=\frac{(c-a)^2}{7}.
$$

Equivalently, since $b-a=2(c-a)$,
$$
\mathrm{Var}(X)=\frac{(b-a)^2}{28}.
$$

If the expected value equals the most likely value, then the model becomes symmetric

$$
\mathrm{Var}_{\text{PERT}}(X)=\frac{(c-a)^2}{7}.
$$
