### 1. Probability Space

The sample space is

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

with sigma-algebra

$$
\mathfrak F=\mathfrak P(\Omega).
$$

The probability measure is

$$
P(\{\omega_1\})=0.50,\quad
P(\{\omega_2\})=0.10,\quad
P(\{\omega_3\})=0.20,\quad
P(\{\omega_4\})=0.20.
$$

Since

$$
0.50+0.10+0.20+0.20=1,
$$

this defines a valid probability space.

---

### 2. Sigma-algebra Generated by $Y$

The random variable $Y$ is

$$
Y(\omega_1)=0,\quad Y(\omega_2)=0,\quad
Y(\omega_3)=1,\quad Y(\omega_4)=1.
$$

Thus,

$$
Y^{-1}(\{0\})=\{\omega_1,\omega_2\}, \quad
Y^{-1}(\{1\})=\{\omega_3,\omega_4\}.
$$

Hence,

$$
\sigma(Y)=\{\varnothing,\Omega,\{\omega_1,\omega_2\},\{\omega_3,\omega_4\}\}.
$$

---

### 3. Measurability of $E[X|Y]$

Since $E[X|Y]$ must be $\sigma(Y)$-measurable, it can depend only on $Y$.

Therefore, it must be constant on:

$$
\{\omega_1,\omega_2\}, \quad \{\omega_3,\omega_4\}.
$$

Let

$$
E[X|Y](\omega)=
\begin{cases}
a, & \omega\in\{\omega_1,\omega_2\},\\
b, & \omega\in\{\omega_3,\omega_4\}.
\end{cases}
$$

---

### 4. Compute $E[X|Y]$

For $\{\omega_1,\omega_2\}$:

$$
0.60a = 0.10 \Rightarrow a = \frac{1}{6}
$$

For $\{\omega_3,\omega_4\}$:

$$
0.40b = 0.20 \Rightarrow b = \frac{1}{2}
$$

Thus,

$$
E[X|Y] = \frac{1}{6}\mathbb{1}_{\{Y=0\}} + \frac{1}{2}\mathbb{1}_{\{Y=1\}}.
$$

---

### 5. Verification

For $A=\{\omega_1,\omega_2\}$:

$$
\int_A E[X|Y] dP = \frac{1}{6}\cdot0.60 = 0.10
$$

$$
\int_A X dP = 0.10
$$

For $A=\{\omega_3,\omega_4\}$:

$$
\int_A E[X|Y] dP = \frac{1}{2}\cdot0.40 = 0.20
$$

$$
\int_A X dP = 0.20
$$

Hence verified.

### Question 2 soloution

To derive the conditional density $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$, we start from the definition of conditional density:

$$
f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})
= \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}.
$$

---

###   Decomposition of the Quadratic Form

From the joint multivariate normal distribution, the exponent can be decomposed using the Schur complement:

$$
Q_{\text{joint}} =
(\mathbf{y} - \boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_y)
+
(\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*)
$$

where

$$
\mathbf{S}_{XX} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX},
$$

and

$$
\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y).
$$

---

###  Marginal Density of $\mathbf{Y}$

The marginal density of $\mathbf{Y}$ has exponent:

$$
Q_{\text{marginal}} =
(\mathbf{y} - \boldsymbol{\mu}_y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_y).
$$

---

### Forming the Conditional Density

Dividing the joint density by the marginal density removes the $\mathbf{y}$-term in the exponent:

$$
Q_{\text{conditional}} = Q_{\text{joint}} - Q_{\text{marginal}}
= (\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*).
$$

---

### Simplifying the Normalizing Constant

The constants are:

- Joint:

$$
\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}}
$$

- Marginal:

$$
\frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}}
$$

Dividing gives:

$$
\frac{1}{\sqrt{(2\pi)^n \cdot \frac{|\boldsymbol{\Sigma}|}{|\boldsymbol{\Sigma}_{YY}|}}}.
$$

Using the determinant identity

$$
|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{YY}| \cdot |\mathbf{S}_{XX}|,
$$

we obtain

$$
\frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}}.
$$

---

### Final Conditional Density

Combining the simplified constant and exponent:

$$
f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})
=
\frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}}
\exp\left(
-\frac{1}{2}
(\mathbf{x} - \boldsymbol{\mu}^*)^T
\mathbf{S}_{XX}^{-1}
(\mathbf{x} - \boldsymbol{\mu}^*)
\right).
$$

### Q1. Joint Distribution of (X, Y)

Since X ~ N(0,1) and Q ~ N(0,4) are independent:

Y = X + Q

Mean:
E[Y] = E[X] + E[Q] = 0 + 0 = 0

Variance:
Var(Y) = Var(X) + Var(Q) = 1 + 4 = 5

Thus:
Y ~ N(0,5)

Covariance:
Cov(X,Y) = E[X(X+Q)] = E[X^2] + E[XQ] = 1 + 0 = 1

Joint distribution:
(X, Y) ~ N( [0,0], [[1,1],[1,5]] )

### Q2. Information Geometry

X is NOT σ(Y)-measurable because Y = X + Q is not invertible.

For a given Y = y, infinitely many (X, Q) pairs satisfy:
x + q = y

So X cannot be written as a function of Y alone.

Practical meaning:
A high power reading (Y) could be due to:
- High temperature (X)
- High mechanical load (Q)

Thus the engineer cannot uniquely determine X from Y.

### Q3. Conditional Expectation

Formula:
E[X | Y=y] = μ_X + (Cov(X,Y)/Var(Y))(y - μ_Y)

Substitute values:
E[X | Y=y] = 0 + (1/5)(y - 0) = y/5

Thus:
E[X | Y] = 0.2Y

Interpretation:
Signal-to-total variance ratio = Var(X)/Var(Y) = 1/5 = 0.2

Only 20% of Y is trusted as temperature signal.

### Q4. Residual Verification

Let:
X̂ = E[X|Y] = 0.2Y
Z = X - X̂ = X - 0.2Y

Mean:
E[Z] = E[X] - 0.2E[Y] = 0 - 0 = 0

Orthogonality:
Cov(Z,Y) = E[(X - 0.2Y)Y] = E[XY] - 0.2E[Y^2]

E[XY] = 1
E[Y^2] = Var(Y) = 5

So:
Cov(Z,Y) = 1 - 0.2(5) = 1 - 1 = 0

Hence Z is uncorrelated with Y.
